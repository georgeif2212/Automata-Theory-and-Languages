## Gramática de libre contexto

Se compone de símbolo inicial, reglas de derivación, variables y símbolos terminales.

Supongamos que queremos sreconocer cadenas dela forma: `0^n1^n`

<center>

`00001111`

`0000011111`

</center>


`S`: símbolo inicial (start)

### Ejemplo de gramática de libre contexto:
`S -> 0S1 | ''` :  (La variable S se puede remplazar por S01 o cadena vacía)

Cadena de entrada: `000111`

El siguiente es un ejemplo de derivación, variables y símbolos terminales:

`S -> 0S1 -> 00S11 -> 000S111 -> 000111`

### Segundo ejemplo de gramática de libre contexto:

Queremos reconocer operaciones de sumas 1+2

`S -> S '+' S | '1' | '2'` (La variable S se puede remplazar por S+S o 1 o 2)

`S -> S + S -> 1 + S -> 1 + 2`

<strong>Alternativamente se pueden usar expresiones regulares para símbolos terminales:</strong>

Si se quisieran reconocer números de más de un dígito, por ejemplo: 11 + 23

`S -> S '+' S | números`

`numeros = [0-9]+`

En el código, el sistema funciona de la siguiente forma:

`11 + 23 => |Analizador léxico (AFD) => token_numero token_suma token_numero => Gramática | => El orden es correcto (reconoce)`

`S -> S '+' S -> 11 + S -> 11 + 23`
